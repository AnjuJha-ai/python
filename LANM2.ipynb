```python
def display_matrix(matrix):
    for row in matrix:
        print(row)

def get_diagonal(matrix):
    return [matrix[i][i] for i in range(min(len(matrix), len(matrix[0])))]

def calculate_trace(matrix):
    return sum(matrix[i][i] for i in range(min(len(matrix), len(matrix[0]))))

def main():
    print("Enter the number of rows and columns for the matrix:")
    rows = int(input("Number of rows: "))
    cols = int(input("Number of columns: "))

    # Input matrix elements from the user
    print("\nEnter the elements of the matrix:")
    matrix = [[float(input(f"Enter element at position ({i + 1},{j + 1}): ")) for j in range(cols)] for i in range(rows)]

    while True:
        # Display menu
        print("\nMenu:")
        print("a) Display Matrix")
        print("b) Display Diagonal")
        print("c) Display Trace")
        print("d) Exit")

        choice = input("Enter your choice (a/b/c/d): ")

        if choice == 'a':
            # Display the matrix
            print("\nMatrix:")
            display_matrix(matrix)
        elif choice == 'b':
            # Display the diagonal of the matrix
            diagonal = get_diagonal(matrix)
            print("\nDiagonal of the Matrix:")
            print(diagonal)
        elif choice == 'c':
            # Display the trace of the matrix
            trace = calculate_trace(matrix)
            print("\nTrace of the Matrix:")
            print(trace)
        elif choice == 'd':
            # Exit the program
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```
