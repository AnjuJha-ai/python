```python
def print_matrix(matrix):
    for row in matrix:
        for element in row:
            print(element, end=' ')
        print()

def get_diagonal(matrix):
    diagonal = []
    for i in range(len(matrix)):
        diagonal.append(matrix[i][i])
    return diagonal

def get_trace(matrix):
    trace = 0
    for i in range(len(matrix)):
        trace += matrix[i][i]
    return trace

def main():
    print("enter no.of rows and coln of matrix:")
    row=int(input("enter no.of row:"))
    col=int(input("enter no.of col:"))
    print("\n enter element of matrix:")
    matrix=[[float(input(f"enter element at position({i+1},{j+1}):"))for j in range(col)]for i in range(row)]
   # matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

    while True:
        print("\nMenu:")
        print("1. Print matrix")
        print("2. Print diagonal")
        print("3. Print trace")
        print("4. Exit")
        choice = int(input("Enter your choice: "))
        if choice == 1:
            print_matrix(matrix)
        elif choice == 2:
            diagonal = get_diagonal(matrix)
            print("Diagonal:", diagonal)
        elif choice == 3:
            trace = get_trace(matrix)
            print("Trace:", trace)
        elif choice == 4:
            break
        else:
            print("Invalid choice")

if __name__ == '__main__':
    main()
```

```python
def display_matrix(matrix):
    for row in matrix:
        print(row)

def get_diagonal(matrix):
    return [matrix[i][i] for i in range(min(len(matrix), len(matrix[0])))]

def calculate_trace(matrix):
    return sum(matrix[i][i] for i in range(min(len(matrix), len(matrix[0]))))

def main():
    print("Enter the number of rows and columns for the matrix:")
    rows = int(input("Number of rows: "))
    cols = int(input("Number of columns: "))

    # Input matrix elements from the user
    print("\nEnter the elements of the matrix:")
    matrix = [[float(input(f"Enter element at position ({i + 1},{j + 1}): ")) for j in range(cols)] for i in range(rows)]

    while True:
        # Display menu
        print("\nMenu:")
        print("a) Display Matrix")
        print("b) Display Diagonal")
        print("c) Display Trace")
        print("d) Exit")

        choice = input("Enter your choice (a/b/c/d): ")

        if choice == 'a':
            # Display the matrix
            print("\nMatrix:")
            display_matrix(matrix)
        elif choice == 'b':
            # Display the diagonal of the matrix
            diagonal = get_diagonal(matrix)
            print("\nDiagonal of the Matrix:")
            print(diagonal)
        elif choice == 'c':
            # Display the trace of the matrix
            trace = calculate_trace(matrix)
            print("\nTrace of the Matrix:")
            print(trace)
        elif choice == 'd':
            # Exit the program
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```
